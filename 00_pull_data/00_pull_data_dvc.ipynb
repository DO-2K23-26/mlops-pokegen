# 00 — Pull des données (S3 via DVC)

**Fin d'étape** : sauvegarde `mlops/artifacts/pipeline/steps/00_pull_data/state.json` (chemin du dataset).

```python
import sys
from pathlib import Path

REPO_ROOT = Path.cwd()
for p in [Path.cwd(), *Path.cwd().parents]:
    if (p / "mlops").is_dir():
        REPO_ROOT = p
        break
sys.path.insert(0, str(REPO_ROOT))

import subprocess
import sys

from mlops.shared.paths import RAW_CARDS_DIR
from mlops.shared.step_artifacts import rel_path, save_step_output

cards_dvc = REPO_ROOT / "data" / "raw" / "cards.dvc"
cards_dir = RAW_CARDS_DIR
print(f"Repo: {REPO_ROOT}")
```

```python
result = subprocess.run(
    ["dvc", "pull", str(cards_dvc)],
    cwd=REPO_ROOT,
    capture_output=True,
    text=True,
)
print(result.stdout)
if result.returncode != 0:
    print(result.stderr, file=sys.stderr)
    raise SystemExit(result.returncode)

n_png = len(list(cards_dir.rglob("*.png")))
print(f"Pull terminé — {n_png} PNG sous {cards_dir}")
```

```python
save_step_output("00_pull_data", {
    "cards_dir": rel_path(cards_dir),
    "n_png": n_png,
    "dvc_file": rel_path(cards_dvc),
})
```
