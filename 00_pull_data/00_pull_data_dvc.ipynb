# 00 — Pull des données (S3 via DVC)

**Fin d'étape** : sauvegarde `artifacts/pipeline/steps/00_pull_data/state.json` (chemin du dataset).

```python
import os
import sys
from pathlib import Path

REPO_ROOT = Path.cwd().resolve()
if not (REPO_ROOT / "shared").exists():
    REPO_ROOT = REPO_ROOT.parent

sys.path.insert(0, str(REPO_ROOT))

from dvc.api import DVCFileSystem

os.environ["AWS_ACCESS_KEY_ID"] = os.getenv("ACCESS_KEY_ID")
os.environ["AWS_SECRET_ACCESS_KEY"] = os.getenv("SECRET_ACCESS_KEY")
os.environ["AWS_DEFAULT_REGION"] = "auto"
print("AWS_ACCESS_KEY_ID:", os.getenv("AWS_ACCESS_KEY_ID"))
print("AWS_SECRET_ACCESS_KEY:", os.getenv("AWS_SECRET_ACCESS_KEY"))

repo = "https://git.razano.dev/llabeyrie/mlops-dataset.git"

fs = DVCFileSystem(repo, rev="main")


```

```python
fs.get("cards", "../data/raw/cards", recursive=True)

n_png = len(list(fs.glob("../data/raw/cards/**/*.png")))

print(f"Pull terminé — {n_png} PNG sous {'../data/raw/cards'}")
```
