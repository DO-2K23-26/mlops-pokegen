# 07 — Monitoring

**Début** : charge `06_deployment` (+ métriques `04`). **Fin** : rapport JSON.

```python
import sys
from pathlib import Path

REPO_ROOT = Path.cwd()
for p in [Path.cwd(), *Path.cwd().parents]:
    if (p / "mlops").is_dir():
        REPO_ROOT = p
        break
sys.path.insert(0, str(REPO_ROOT))

import json

from mlops.shared.paths import STEP_07_REPORT
from mlops.shared.step_artifacts import load_previous_step_output, load_step_output, rel_path, save_step_output

prev = load_previous_step_output("07_monitoring")
eval_state = load_step_output("04_evaluation")
train_state = load_step_output("03_train_model")

report = {
    "deployment_samples_dir": prev.get("deployment_samples_dir"),
    "sample_paths": prev.get("sample_paths", []),
    "final_train_loss": train_state.get("final_train_loss"),
    "final_val_loss": eval_state.get("final_val_loss") or train_state.get("final_val_loss"),
    "epochs_done": train_state.get("epochs_done"),
    "lora_dir": prev.get("lora_dir"),
}
print(json.dumps(report, indent=2))

STEP_07_REPORT.parent.mkdir(parents=True, exist_ok=True)
STEP_07_REPORT.write_text(json.dumps(report, indent=2), encoding="utf-8")

save_step_output("07_monitoring", {
    "report_path": rel_path(STEP_07_REPORT),
    **report,
})
print(f"Rapport → {STEP_07_REPORT}")
```
